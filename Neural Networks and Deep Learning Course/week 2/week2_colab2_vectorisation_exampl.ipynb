```python
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

start = time.time()
result = a.dot(b)
end = time.time()

print("result vectorisation :", result)
print("vectorisation time   :" + str(end - start) + "ms")

start = time.time()
result = 0
for i in range(1000000):
  result += a[i] * b[i]
end = time.time()

print("result loop :", result)
print("loop time   :" + str(end - start) + "ms")
```
